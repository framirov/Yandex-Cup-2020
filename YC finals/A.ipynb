```python
import pandas as pd
import numpy as np
```

```python
heats=pd.read_csv('heats.csv')
heats#.head()
```

```python
zapliv=[[0 for x in range(8)] for x in range(3)]
speed = [4,5,3,6,2,7,1,8]
#zapliv
speed.reverse()
print(speed)
```

```python
heats_sort=heats.copy()
heats_sort=heats_sort.sort_values('time')
heats_sort.index=(range(len(heats_sort)))
heats_sort
```

```python
sch=len(heats_sort)
A=[]
for n,z in enumerate(zapliv):
    for i in speed:
        sch=sch-1
        time=heats_sort.loc[sch,'time']
        id=heats_sort.loc[sch,'id']
        z[i-1]=(id,(n+1),i, time)
        A.append((id,(n+1),i))
        
    print(z)
    print(' ')
```

```python
print(A)
```

```python
answer=[0 for x in range(8)]
for a in A:
    if a[0]>300:
        answer[a[0]-301]=a
for a in answer:
    print(f'{a[0]},{a[1]},{a[2]}')
```
