```python
def get_limit(x):
    return(limits[x])

n, k = [int(x) for x in input().strip().split(' ')]
line = input().strip().split(' ')
third = sorted([int(x) for x in input().strip().split(' ')])
zv_even=[]
zv_odd=[]
limits={}
fl = True
ans = ''
if (k == n):
    if (n % 2 != 0):
        smalls = third[:(k//2 + 1)]
        bigs = third[(k//2 + 1):]
        for i in range(n):
            if (i % 2 == 0):
                ans += str(smalls[0]) + ' '
                smalls.pop(0)
            else:
                ans += str(bigs[0]) + ' '
                bigs.pop(0)
    else:
        smalls = third[:(k//2)]
        bigs = third[(k//2):]
        for i in range(n):
            if (i % 2 == 0):
                ans += str(smalls[0]) + ' '
                smalls.pop(0)
            else:
                ans += str(bigs[0]) + ' '
                bigs.pop(0)        
    print(ans[:len(ans)-1])
else:
    prev = k
    while k > 0:
        for i,y in enumerate(line):
            if y=='-':
                if (i == 0):
                    if (line[i+1] == '-'):
                        continue
                    else:
                        if (i not in zv_even):
                            zv_even.append(i)
                        limits[i] = int(line[i+1])
                elif (i == n-1):
                    if (line[i-1] == '-'):
                        continue
                    else:
                        if i%2 == 0:
                            if (i not in zv_even):
                                zv_even.append(i)
                            limits[i] = int(line[i-1])
                        else:
                            if (i not in zv_odd):
                                zv_odd.append(i)
                            limits[i] = int(line[i-1])
                elif (line[i-1] == '-'):
                    if line[i+1] == '-':
                        continue
                    else:
                        if i%2 ==0:
                            if (i not in zv_even):
                                zv_even.append(i)
                            limits[i]=int(line[i+1])
                        else:
                            if (i not in zv_odd):
                                zv_odd.append(i)
                            limits[i]=int(line[i+1])
                elif (line[i+1] == '-'):
                    if line[i-1] == '-':
                        continue
                    else:
                        if i%2 == 0:
                            if (i not in zv_even):
                                zv_even.append(i)
                            limits[i]=int(line[i-1])
                        else:
                            if (i not in zv_odd):
                                zv_odd.append(i)
                            limits[i]=int(line[i-1])
                else:
                    if i%2 ==0:
                        if (i not in zv_even):
                            zv_even.append(i)
                        limits[i]=min(int(line[i-1]), int(line[i+1]))
                    else:
                        if (i not in zv_odd):
                            zv_odd.append(i)
                        limits[i]=max(int(line[i-1]), int(line[i+1]))
        zv_even=sorted(zv_even, key = get_limit)
        zv_odd=sorted(zv_odd, key = get_limit, reverse = True)
        while (len(zv_odd) > 0):
            if((third == []) |(limits[zv_odd[0]] > third[len(third) - 1])):
                break
            line[zv_odd[0]] = str(third[len(third) - 1])
            try:
                limits[i+1] = min(limits[i+1], third[len(third) - 1])
            except:
                pass
            try:
                limits[i-1] = min(limits[i-1], third[len(third) - 1])
            except:
                pass
            third.pop(len(third) - 1)
            k -= 1
            zv_odd.pop(0)
        while (len(zv_even) > 0):
            if((third == []) | (limits[zv_even[0]] < third[0])):
                break
            line[zv_even[0]] = str(third[0])
            try:
                limits[i+1] = max(limits[i+1], third[0])
            except:
                pass
            try:
                limits[i-1] = max(limits[i-1], third[0])
            except:
                pass
            third.pop(0)
            k -= 1
            zv_even.pop(0)
        if (k == prev):
            fl = False
            break
        prev = k
    if (fl):
        for i in line:
            ans += str(i) + ' '
        ans = ans[:len(ans)-1]
        print(ans)
    else:
        print("NO SOLUTION")
```
