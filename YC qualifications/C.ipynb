```python
import csv
logs = []
with open('logs.csv', newline='') as File:  
    reader = csv.reader(File)
    for row in reader:
        logs += [row]
was = set()
n_to_adress = {}
adress_to_n = {}
count = {}
x = 0
for line in logs:
    if (line[0] == ''):
        if (line[1] in was):
            count[adress_to_n[line[1]]] += 1
        else:
            was.add(line[1])
            n_to_adress[x] = set()
            n_to_adress[x].add(line[1])
            adress_to_n[line[1]] = x
            count[x] = 1
            x +=1
    elif (line[1] == ''):
        if (line[0] in was):
            count[adress_to_n[line[0]]] += 1
        else:
            was.add(line[0])
            n_to_adress[x] = set()
            n_to_adress[x].add(line[0])
            adress_to_n[line[0]] = x
            count[x] = 1
            x +=1
    if (line[0] not in was) & (line[1] not in was):
            was.add(line[0])
            was.add(line[1])
            n_to_adress[x] = set()
            n_to_adress[x].add(line[0])
            n_to_adress[x].add(line[1])
            adress_to_n[line[0]] = x
            adress_to_n[line[1]] = x
            count[x] = 1
            x += 1
    elif ((line[0] in was) & (line[1] in was)):
        if (adress_to_n[line[0]] == adress_to_n[line[1]]):
            count[adress_to_n[line[0]]] += 1
        else:
            temp_n = adress_to_n[line[1]]
            temp_adresses = n_to_adress[adress_to_n[line[1]]]
            count[adress_to_n[line[0]]] += count[adress_to_n[line[1]]] + 1
            for lines in temp_adresses:
                n_to_adress[adress_to_n[line[0]]].add(lines)
                adress_to_n[lines] = adress_to_n[line[0]]
            n_to_adress[temp_n] = []
            count[temp_n] = 0
    elif ((line[0] in was) & (line[1] not in was)):
        was.add(line[1])
        adress_to_n[line[1]] = adress_to_n[line[0]]
        n_to_adress[adress_to_n[line[0]]].add(line[1])
        count[adress_to_n[line[0]]] += 1
    elif ((line[0] not in was) & (line[1] in was)):
        was.add(line[0])
        adress_to_n[line[0]] = adress_to_n[line[1]]
        n_to_adress[adress_to_n[line[1]]].add(line[0])
        count[adress_to_n[line[1]]] += 1
maxx = 0
for i in range(x):
    if count[i] > maxx:
        maxx = count[i]
        maxn = i
print(n_to_adress[maxn])
print(maxx)
```
