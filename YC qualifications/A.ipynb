```python
import pandas as pd
import numpy as np
from datetime import datetime,timedelta
```

```python
logs=pd.read_csv('log.csv')#, dtype={'date': pd.datetime})
logs['date']=pd.to_datetime(logs['date'].apply(lambda x: x.replace('_',' ')))
logs
```

```python
#first_task
log_0414=logs[(logs['date'].dt.date==datetime(2020,4,11).date())]
users=log_0414['user'].unique()
log_0414=log_0414.sort_values(['date'])
session=len(log_0414)
for user in users:
    #print(log_0414[log_0414['user']==user]['date'])
    #print((log_0414[log_0414['user']==user]['date'].diff()>timedelta(minutes=30)))
    for i, y in enumerate(log_0414[log_0414['user']==user]['date'].diff()<timedelta(minutes=30)):
        if y:
            print(i)
            session=session-1
            
    
print('First answer is:')
print(session)
print(len(log_0414))
```

```python
#second_task
logs_music=logs[(logs['event_type']==3) & (logs['parameter']=='picture')]

q=0
for d in range(1,31):
    td=datetime(2020,4,d)
    #print(td)
    print(len(logs_music[(logs_music['date'].dt.date==td.date())]['user']))
    w = len(logs_music[(logs_music['date'].dt.date==td.date())]['user'].unique())
    print(td,w)
    if w>q:
        q=w
        #print(td,w)

print('Second answer is:')
print(q)
```
